```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from retinaface import RetinaFace

fig, axs = plt.subplots(4, 6, figsize=(16, 16))

for i in range(4):
    # Load image
    img = mpimg.imread("woong{}.jpg".format(i))

    # Extract faces using RetinaFace
    faces = RetinaFace.extract_faces(img_path="woong{}.jpg".format(i), align=True)

    # Show original image
    axs[i, 0].imshow(img)
    axs[i, 0].set_title("woong{}.jpg".format(i))

    # Show extracted faces
    for j, face in enumerate(faces):
        axs[i, j+1].imshow(face)
        axs[i, j+1].set_title("Face {}".format(j+1))

# Hide axis ticks and labels
for ax in axs.flatten():
    ax.axis("off")

plt.tight_layout()
plt.show()
```
